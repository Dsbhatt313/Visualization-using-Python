Scatter Plot

```python
df_real_estate = pd.read_csv("scatter_data.csv")
```

```python
df_real_estate
```

```python
#Scatter Plot with matplotlib.pyplot library

plt.figure(figsize = (12, 6))
scatter = plt.scatter(df_real_estate["Area (ft.)"],
            df_real_estate["Price"],
            alpha = 0.6,    # for point transparency (range btw 0.0 - 1.0)
            c = df_real_estate["Building Type"],
            cmap = 'viridis')
plt.xlabel("Area (sq. ft.)", weight = 'bold')
plt.ylabel("Price in (000' of $)", weight = 'bold')
plt.legend(*scatter.legend_elements(), 
           loc = "upper left", 
           title = "Building Types")
plt.title("Area vs Price of Califonia Real Estate", 
          fontsize = 16, 
          weight = 'bold', 
          pad = 15)
plt.show()
```

```python
#Scatter Plot with seaborn library

plt.figure(figsize = (12, 6))
sns.scatterplot(x = df_real_estate["Price"], 
                y = df_real_estate["Area (ft.)"],
                hue = df_real_estate["Building Type"], 
                palette = ["black", "darkblue", "purple", "pink", "white"],
                s = 100)
plt.xlabel("Area (sq. ft.)", weight = 'bold')
plt.ylabel("Price in (000' of $)", weight = 'bold')
plt.title("Area vs Price of Califonia Real Estate", 
          fontsize = 16, 
          weight = 'bold', 
          pad = 15)
plt.show()
```

Regression Line Scatter PLot

```python
df_expenditure = pd.read_csv("scatter_plot_ii.csv")
```

```python
df_expenditure
```

```python
#With the help of sns.regplot()

plt.figure(figsize = (12, 6))    #enlarging the graph/figure with matplotlib
#sns.set(rc = {'figure.figsize': (12, 6)})    #enlarging the graph/figure with seaborn
sns.regplot(x = df_expenditure["Budget"],
            y = df_expenditure["Sales"],
            data = df_expenditure,
            scatter_kws = {'color': 'k'},
            line_kws = {'color': 'red'})
plt.xlabel("Budget in (000's of $)")
plt.ylabel("Sales in (000's of units)")
plt.title("Effects of Ad Expenditure on Sales", fontsize = 16, weight = 'bold', pad = 15)
plt.show()
```

```python
#With the help of sns.lmplot()

sns.lmplot(x = "Budget",
           y = "Sales",
           data = df_expenditure, 
           height = 10,    #enlarging the graph/figure
           scatter_kws = {'color': 'k'},
           line_kws = {'color': 'red'})
plt.xlabel("Budget in (000's of $)")
plt.ylabel("Sales in (000's of units)")
plt.title("Effects of Ad Expenditure on Sales", fontsize = 16, weight = 'bold', pad = 15)
plt.show()
```

Bar and Line Chart

```python
from matplotlib.ticker import PercentFormatter    # to format numbers to percentage 
```

```python
df_kdnuggets = pd.read_csv("bar_line_chart_data.csv")
```

```python
df_kdnuggets
```

```python
sns.set_style("white")
fig, ax = plt.subplots(figsize = (10, 6))
ax.bar(x = df_kdnuggets["Year"],
       height = df_kdnuggets["Participants"],
       color = 'k')
ax.set_ylabel("No of Participants",
              weight = 'bold')
ax.tick_params(labelcolor = 'k',
               width = 2,
               labelsize = 'large')
ax1 = ax.twinx()    #twinx() allows for 2 seperate charts to share same x-axis
ax1.set_ylim(0, 1)    #set_ylim() lets us alter the y-axis range by adding a min and max value
ax1.yaxis.set_major_formatter(PercentFormatter(xmax = 1.0))
ax1.plot(df_kdnuggets["Year"],
         df_kdnuggets["Python Users"],
         color = '#b60000',
         marker = "D")
ax1.set_ylabel("Python Users in %",
               weight = 'bold')
ax1.tick_params(labelcolor = '#b60000',
                width = 2,
                labelsize = 'large')
#ax.legend(labels = "Participants", loc = 2)
#ax1.legend(labels = "Python Users", loc = 1)
ax.set_title("KD Nugget Survey Python Users (2012 - 2019)",
         fontsize = 16,
         weight = 'bold',
         pad = 15)
plt.show()
```
